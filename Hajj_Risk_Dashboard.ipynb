```python
# Step 2: Load the data
import pandas as pd

# Replace this with the path to your data file
file_path = r'C:\Users\LOLOY\hajj_pilgrims_final_modified.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Step 3: Build and Run the Dash Application
from jupyter_dash import JupyterDash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px

# Create the Dash app
app = JupyterDash(__name__)

# Extract unique routes
routes = data['Route'].unique()

# Define the layout
app.layout = html.Div([
    html.H1("لوحة تحكم مخاطر طرق الحجاج", style={'textAlign': 'center'}),
    dcc.Dropdown(
        id='route-dropdown',
        options=[{'label': route, 'value': route} for route in routes],
        value=routes[0],
        placeholder="اختر طريقًا"
    ),
    dcc.Graph(id='risk-map')
])

# Define the callback
@app.callback(
    Output('risk-map', 'figure'),
    [Input('route-dropdown', 'value')]
)
def update_map(selected_route):
    # Filter data by selected route
    filtered_data = data[data['Route'] == selected_route]

    # Create a map
    fig = px.scatter_mapbox(
        filtered_data,
        lat='Latitude',  # Replace with the appropriate column
        lon='Longitude',  # Replace with the appropriate column
        color='Risk Level',
        size='Population Density',
        hover_data=['Health Condition', 'Temperature'],
        title=f"خريطة المخاطر للطريق: {selected_route}",
        mapbox_style="carto-positron",
        zoom=10
    )
    return fig

# Run the app
app.run_server(mode='inline')
```
